## Titanic Example
We will first do a simple MVP logistic regression model as shown in the Kernel of Pedro Marcelino; then we will apply a deep learning algorithm to see how much better we get. In a third step we will go through all the manual data analysis steps and classical feature abstraction. I'm curious to see if this results in a better prediction than the deep learning algorithm (and at the same time teaches me a bit of classical ML and the use of pandas / seaborn).

````python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# import scikit learn libraries
from sklearn.model_selection import train_test_split
from sklearn.model_selection import learning_curve
from sklearn.model_selection import validation_curve
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
````

````python
# add support functions
def draw_missing_data_table(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count().sort_values(ascending=False)).round(2)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'], sort=True)
    return missing_data
````

````python
# plot learning curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                       n_jobs=1, train_sizes=np.linspace(0.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel('Training examples')
    plt.ylabel('Score')
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs = n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std, alpha=0.1, color='r')
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std, alpha=0.1, color='g')
    plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training score')
    plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Test score')
    plt.legend(loc='best')
    
    return plt
````

````python
# plot validation curve
def plot_validation_curve(estimator, title, X, y, param_name, param_range, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    train_scores, test_scores = validation_curve(estimator, X, y, param_name, param_range, cv)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    plt.plot(param_range, train_mean, color='r', marker='o', markersize=5, label='Training score')
    plt.fill_between(param_range, train_mean + train_std, train_mean - train_std, alpha=0.15, color='r')
    plt.plot(param_range, test_mean, color='g', linestyle='--', marker='s', markersize=5, label='Validation score')
    plt.fill_between(param_range, test_mean + test_std, test_mean - test_std, alpha=0.15, color='g')
    plt.grid()
    plt.xscale('log')
    plt.legend(loc='best')
    plt.xlabel('Parameter')
    plt.ylabel('Score')
    plt.ylim(ylim)
````

### Initial Data Analysis

````python
df = pd.read_csv(r'C:\Users\Christian\git\dsr\Data\titanic.csv')
df.head()
````

#### First thoughts
Let's have a look at the features first:
* **PassengerId:** The unique identifier. Should not have any impact on the model.
* **Survived:** The target value.
* **Pclass:** The passenger class. Experience shows that rich people typically have it better.
* **Name:** In principle the name should not have any influence on survival. However, we see that there is a title embedded, and this might give us a hint as well.
* **Sex:** The fair sex should have a higher survival rate.
* **Age:** It is still open if the age has a clear impact. Conventional wisdom would say that people would try to save children first.
* **SibSp:** The number of siblings and spouses a person has on board. Could indicate a probability of higher survival as a family can work together.
* **Parch:** Same as above, just with parents and children.
* **Ticket:** Confuse. This should not have an impact.
* **Fare:** Similar to the passenger class - the more someone pays, the higher the likelihood of survival we would assume.
* **Cabin:** There seem to be quite many missing values.
* **Embarked:** Port of embarkation; we would assume that this has no impact on survival rates.

````python
df.describe().round(2)
````

It seems the age is missing for quite a few of the records. The rest seems resonable, with quite high variations in the fare. Let's have another look at the data.

````python
draw_missing_data_table(df)
````

#### Analysis of missing values
* We see that _Cabin_ is empty for more than 75%, so we will remove this feature.
* As we assume that the _Age_ has an impact on survival we will need to find a creative solution. For now we just add dummy values.
* For the _Embarked_ we will just remove the two rows which have no embarkation location.


````python
# drop cabin
df.drop('Cabin', axis=1, inplace=True)

# replace NaN age with dummy value
df['Age'].fillna(1000, inplace=True)

# remove empty embarked records
df.drop(df[pd.isnull(df['Embarked'])].index, inplace=True)

df.head()
````

### Minimum Viable Model (MVM) with Logistic Regression

* Let's drop the _PassengerId_ as it should not have any impact
* The features _Sex_ and _Embarked_ should be categorical
* Featires _Name_ and _Ticket_ are too complicated for the MVM, so we will drop it
* For the sake of simplicity we group _SibSp_ and _Parch_ into one feature _FamilySize_

````python
df.drop('PassengerId', axis=1, inplace=True)
df['Sex'] = pd.Categorical(df['Sex'])
df['Embarked'] = pd.Categorical(df['Embarked'])
df.drop('Name', axis=1, inplace=True)
df.drop('Ticket', axis=1, inplace=True)
df['FamilySize'] = df['SibSp'] + df['Parch']
df.drop('SibSp', axis=1, inplace=True)
df.drop('Parch', axis=1, inplace=True)

df.head()
````

We now need to do the one-hot-encoding. Fortunately this is very easy with ```pandas```.

````python
df = pd.get_dummies(df, drop_first=True)
df.head()
````

Now let's create the train and test data set including the labels:

````python
X = df.loc[:, df.columns != 'Survived']
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
````

Now we come to the last step - fitting the logistic regression.

````python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
````

We now check the model performance using the mean and standard deviation and the learning and validation curves.

````python
scores = cross_val_score(logreg, X_train, y_train, cv=10)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
````

````python
title = 'Learning Curves (Logistic Regression)'
cv = 10
plot_learning_curve(logreg, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, n_jobs=1);
````

We see that the model does not overfit (training score and test score are converging). Unfortunately, the model underfits, and we only have an accuracy below 80%. Let's look at the validation curve which allows us to see how far we could go with hyperparameter optimization. 

````python
title = 'Validation Curve (Logistic Regression)'
param_name = 'C'
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
cv = 10
plot_validation_curve(estimator=logreg, title=title, X=X_train, y=y_train, 
                      param_name=param_name, ylim=(0.5, 1.01), param_range=param_range);
````

The logistic regression has only one parameter (C) which can be tuned. it seems that the optimal value is around 10. Again, we do not see any signs of overfitting, but the model still underfits. Let's now try the same thing with a deep learning algorithm to see if this gives us a significant improvement.

### Deep Learning MVP
In order to apply a deep learning classification algorithm we need to massage the data format slightly to match the expected input.

````python
X_train.head()
#, X_test, y_train, y_test
````

The features _Pclass_, _Age_, _Fare_ and _FamilySize_ are not between 0 and 1. Let's correct that by just dividing them through the largest value in the overall set.

````python
df.describe()
````

````python
X_train.head()
````

````python
pd.options.mode.chained_assignment = None # we should remove this eventually; when I execute the code below it also
                                          # affects X_train / X_test for some reason. See here for more information:
                                          # https://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas
                                          # we could also use the .copy() method
        
X_train_norm = X_train
X_test_norm = X_test

X_train_norm['Pclass'] = X_train['Pclass']/df['Pclass'].max()
X_train_norm['Age'] = X_train['Age']/df['Age'].max()
X_train_norm['Fare'] = X_train['Fare']/df['Fare'].max()
X_train_norm['FamilySize'] = X_train['FamilySize']/df['FamilySize'].max()

X_test_norm['Pclass'] = X_test['Pclass']/df['Pclass'].max()
X_test_norm['Age'] = X_test['Age']/df['Age'].max()
X_test_norm['Fare'] = X_test['Fare']/df['Fare'].max()
X_test_norm['FamilySize'] = X_test['FamilySize']/df['FamilySize'].max()

X_train_norm.head()
````

````python
X_train_norm.describe()
````

````python
X_train_norm.shape
````

````python
from keras import models, layers

model = models.Sequential()
model.add(layers.Dense(
    16,
    activation="relu",
    input_shape=(7, )))
model.add(layers.Dense(16, activation="relu"))
model.add(layers.Dense(1, activation="sigmoid"))

model.summary()

model.compile(
    optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

````python
history = model.fit(
    X_train_norm,
    y_train,
    epochs = 100,
    batch_size=30,
    validation_data=(X_test_norm, y_test),
    verbose=0
)
````

````python
import matplotlib.pyplot as plt

r = range(1, len(history.history["acc"]) + 1)

plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(r, history.history["acc"], label="acc")
plt.plot(r, history.history["val_acc"], label="val_acc")
plt.legend()
plt.title("Accuracy")

plt.subplot(1, 2, 2)
plt.plot(r, history.history["loss"], label="loss")
plt.plot(r, history.history["val_loss"], label="val_loss")
plt.legend()
plt.title("Loss")

plt.show()
plt.close()
````

It seems that the deep learning algorithm is not really better than our simple logistic regression. Let's try both again after our manual feature engineering.

````python
df['Embarked'] = pd.Categorical(df['Embarked'])
df['Pclass'] = pd.Categorical(df['Embarked'])
df.dtypes
````

````python
df = pd.get_dummies(df, drop_first=1)
df.head()
````

````python
from sklearn.model_selection import train_test_split
X = df.loc[:,df.columns != 'Survived']
y = df.loc[:,df.columns == 'Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

Now we need to normalize the _Fare_ feature. For this we use a Box-Cox which normalizes a feature optimally.

````python
from scipy.stats import boxcox
X_train_transformed = X_train.copy()
X_train_transformed['Fare'] = boxcox(X_train_transformed['Fare'] + 1)[0]
X_test_transformed = X_test.copy()
X_test_transformed['Fare'] = boxcox(X_test_transformed['Fare'] + 1)[0]
X_train_transformed.describe()
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_transformed_scaled = scaler.fit_transform(X_train_transformed)
X_test_transformed_scaled = scaler.fit_transform(X_test_transformed)
````

Get polynomial features. To me it's not so clear what we are doing here.

````python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2).fit(X_train_transformed)
X_train_poly = poly.transform(X_train_transformed_scaled)
X_test_poly = poly.transform(X_test_transformed_scaled)
print(poly.get_feature_names())
````

````python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

y_train = np.ravel(y_train)

# get score of original model
logreg = LogisticRegression(C=1)
logreg.fit(X_train, y_train)
scores = cross_val_score(logreg, X_train, y_train, cv=10)
print('CV accuracy (original): %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
highest_score = np.mean(scores)

# get score of model with feature selection (for each number of features)
for i in range(1, X_train_poly.shape[1]+1, 1):

    # select i features
    select = SelectKBest(score_func=chi2, k=i)
    select.fit(X_train_poly, y_train)
    X_train_poly_selected = select.transform(X_train_poly)

    # model with i features selected
    logreg.fit(X_train_poly_selected, y_train)
    scores = cross_val_score(logreg, X_train_poly_selected, y_train, cv=10)
    print('CV accuracy (number of features = %i): %.3f +/- %.3f' % (i, np.mean(scores), np.std(scores)))
    
    # save highest score
    if np.mean(scores) > highest_score:
        highest_score = np.mean(scores)
        std = np.std(scores)
        k_features_highest_score = i
    elif np.mean(scores) == highest_score:
        if np.std(scores) < std:
            std = np.std(scores)
            k_features_highest_score = i
            
print('Number of features when highest score: %i' % k_features_highest_score)    
````

Now that we know the number of (polynomial) features we can go and fit the model and check the learning and validation curves. 

````python
select = SelectKBest(score_func=chi2, k=k_features_highest_score)
select.fit(X_train_poly, y_train)
X_train_poly_selected = select.transform(X_train_poly)
````

````python
logreg = LogisticRegression(C=1)
logreg.fit(X_train_poly_selected, y_train)
````

````python
scores = cross_val_score(logreg, X_train_poly_selected, y_train, cv=10)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
````

````python
title = "Learning Curves (Logistic Regression)"
cv = 10
plot_learning_curve(logreg, title, X_train_poly_selected, 
                    y_train, ylim=(0.7, 1.01), cv=cv, n_jobs=1);
````

````python
title = 'Validation Curve (Logistic Regression)'
param_name = 'C'
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0] 
cv = 10
plot_validation_curve(estimator=logreg, title=title, X=X_train_poly_selected, y=y_train, 
                      param_name=param_name, ylim=(0.5, 1.01), param_range=param_range);
````
