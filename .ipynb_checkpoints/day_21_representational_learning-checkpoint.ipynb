### Representational learning:
* When we want to find pictures similar to existing pictures: we can't just take a distance measures on the pixels (e.g. this would not work in case of a rotated picture)
* Instead, we could use convolutional NN
* CNN reduces size, but increase features

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from keras import models, layers
from keras import applications
```

```python
vgg19 = applications.VGG19(include_top=False, input_shape=(224, 224, 3)) # here 'top' means what we normally called bottom
                                                                         # basically it's equivalent to three times vgg19.pop()
vgg19.summary()
```

```python
224 * 244 * 3 / 25088
```

The model compresses the pictures by a factor of about 6 as the latent vector has only 25'088 values.

We can download the VGG19 and then throw away the dense layers in order to re-train the model (but keep all the feature extraction as there is where the main work is done - and already was done).

For all our reference pictures we create the flattened version (i.e. the one with the representation in a 25'088 element vector).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import glob2 as glob

#glob_search_path = 'data\\cats_vs_dogs_small\\train\\**\\*.jpg'
glob_search_path = 'data/cats_vs_dogs_small/train/**/*.jpg'
paths = glob.glob(glob_search_path)
print('\n'.join(paths))
```

Let's look at one of these cats and dogs:

```python
image_cat = mpimg.imread(path)
plt.imshow(image_cat, cmap='gray')
plt.show()
plt.close()
```

```python
from PIL import Image

pic = Image.open(path)
pic = np.array(pic)
print(type(pic), pic.shape)
```

Unfortunately the image is not in the required format of 224 x 224, so we need to resize. We could do this with numpy:

```python
pic = np.resize(pic, (224, 224, 3))
plt.imshow(pic, cmap='gray')
plt.show()
plt.close()
```

Resizing in numpy does not work well it seems (it doesn't consider image characteristics). So let's do the same again in PIL.

```python
pic = Image.open(path)
pic.thumbnail((224, 224), Image.ANTIALIAS)
pic = np.array(pic)

print(type(pic), pic.shape)
plt.imshow(pic, cmap='gray')
plt.show()
plt.close()
```

Still not good as it preserves the proportions.

```python
pic = Image.open(path)
pic = pic.resize((224, 224))
pic = np.array(pic)

print(type(pic), pic.shape)
plt.imshow(pic, cmap='gray')
plt.show()
plt.close()
```

Now we have it nicely squeezed.

```python
latent_vec = vgg19.predict(pic.reshape((1, 224, 224, 3))) # in keras, the prediction works on batches of images, thus we have 
                                                          # to add the '1' as the first dimension.

print(latent_vec.flatten())
```
