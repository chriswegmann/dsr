### Classical Machine Learning

We can define classical ML as everything except NN with one or more hidden layers. Linear regression is also ML and embedded in Excel; it will be a question of time that also random forest etc. will be embedded. When comparing models it's about different predictions, complexity and speed (and other aspects).

NN often take on the burden of feature engineering (example of iris tabular data vs. iris pictures). The architecture (i.e. which models are stacked on top of each other in which order) is still kind of a black art. It's all hypothesis-based and iterative so far.

### Introduction

In _decision trees_, we construct trees from the root and then sequentially add branches (=nodes) until we can't branch anymore. At a given branch, we select the feature that maximizes the information (= reduces the uncertainty). For regression problems, we typically use residual sum of squeares as a measure; for classification problems we use entropy. Please see [here](https://www.slideshare.net/marinasantini1/lecture-4-decision-trees-2-entropy-information-gain-gain-ratio-55241087) for more details on definitions of information.

We then can create _bagged trees_ instead of a single decision tree. Here we first use only a subset of our data (typically 2/3) and a subset of our features (typically the square root), but instead generate many trees. We sample both data and features with replacement. Once we have generated all trees (according to the algorithm laid out above) we take the average of all trees and typically come up with a better result than one single tree over all features.

The bagged trees approach has disadvantages because the trees can be heavily correlated. This happens if we have a couple of strongly predictive features in our data set, so the individual trees would always pick the same or similar features as their top nodes. To prevent this, we use a technique called _random forest_ which restrict the features available for selection at a given branch randomly. Like this we de-correlate the ensemble of trees and typically reduce the variance across all trees.

See also [here](https://towardsdatascience.com/why-random-forest-is-my-favorite-machine-learning-model-b97651fa3706) for a longer description of random forests.

Disadvantages of decision trees:
* Splits space into rectangles
* Tends to overfit
* Is not robust if new data gets added
* Almost never the optimal predictive power

We can improve this with _Bagging (= B-ootstrapping and Agg-regat-ing)_. The idea is that an ensemble of trees in general is better than an individual tree. However, this is not restricted to decision trees - we can also have e.g. ensembles of NN.

### Random Forest: Distributed Grid Points
See definition in slides.
We want to have individual trees that are strong, but not correlated.
Trees are independent, i.e. we can parallelize them.

Despite out-of-bag validation: we still need k folds with validations, and we still need a train and test split.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
x, y = make_moons(n_samples=1000, noise=0.35, random_state=2) # this generates points on a grid which are distributed
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=42)

x_vis = x_train.transpose()
sns.scatterplot(x_vis[0], x_vis[1], y_train);
````

````python
rf_grid_points = RandomForestClassifier(n_estimators=5, random_state=2)
rf_grid_points.fit(x_train, y_train)
y_pred = rf_grid_points.predict(x_test)
````

````python
fig, axs = plt.subplots(ncols=2, figsize=(20,5))

x_vis = x_test.transpose()
sns.scatterplot(x_vis[0], x_vis[1], y_test, ax=axs[0]).set_title("True values");
sns.scatterplot(x_vis[0], x_vis[1], y_pred, ax=axs[1]).set_title("Predicted values");
````

````python
def show_accuracy(x_test, y_test, model):
    y_pred = model.predict(x_test)
    print("False positives: " + str(sum(pd.DataFrame(y_test - y_pred)[0]==-1)))
    print("False negatives: " + str(sum(pd.DataFrame(y_test - y_pred)[0]==1)))
    print("Accuracy:        " + str(100 * round(model.score(x_test, y_test), 2)) + "%")
````

````python
show_accuracy(x_test, y_test, rf_grid_points)
````

### Random Forest: Cancer Images

````python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
````

````python
pd.DataFrame(cancer.data, columns=cancer.feature_names).head()
````

````python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
````

````python
forest_cancer = RandomForestClassifier(n_estimators=5, random_state=2)
forest_cancer.fit(x_train, y_train)
````

````python
show_accuracy(x_test, y_test, forest_cancer)
````

````python
# store feature importance
feat_imp_forest_cancer = pd.DataFrame.from_dict(dict(zip(cancer.feature_names, forest_cancer.feature_importances_)), 
                                             orient='index', 
                                             columns=['rand_for'])
````

### Decision Tree: Cancer Images

````python
dec_tree_cancer = DecisionTreeClassifier()
dec_tree_cancer.fit(x_train, y_train)
````

````python
show_accuracy(x_test, y_test, dec_tree_cancer)
````

````python
# store feature importance
feat_imp_dec_tree_cancer = pd.DataFrame.from_dict(dict(zip(cancer.feature_names, dec_tree_cancer.feature_importances_)), 
                                           orient='index', 
                                           columns=['dec_tree'])
````

````python
# compare feature importance
feat_imp = pd.concat([feat_imp_forest_cancer, feat_imp_dec_tree_cancer], axis=1, sort=False)
feat_imp = feat_imp.sort_values('rand_for', ascending=False)
feat_imp = feat_imp.rank(method='max')
feat_imp
````

Slide with bias / variance
* Blue curve has high variance, low bias
* Red lines has high bias, low variance

Everything with high variance is overfitted. We typically start with overfitted models and then gradually reduce the model complexity. We know that we have bias if the model is too simple and it's not predicting well. It's difficult to correct because we don't know what the 'center' is.

ROC curve is the preferred method to evaluate binary classifiers. This is used e.g. as a criteria in a grid search.

### Test ROC

````python
import numpy as np

def gen_data(n):
    x_1 = np.random.rand(n, 5)
    x_2 = np.random.rand(n, 5) * 0.75
    x = np.concatenate((x_1, x_2), axis=1)
    y = x[:,0]*np.sin(x[:,0]) + np.sin(2*x[:,1]) + 3*x[:,2] + .4*x[:,3] + x[:,4]
    
    return train_test_split(x, y, test_size=0.5)

x_train, x_test, y_train, y_test = gen_data(100)
````

````python
# fit random forest to train data
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=5, random_state=2)
forest.fit(x_train, y_train)

# show accuracy
print("Accuracy:        " + str(100 * round(forest.score(x_test, y_test), 2)) + "%")
````

````python
# store feature importance
feat_imp_val = pd.DataFrame.from_dict(dict(zip(range(10), forest.feature_importances_)), 
                                             orient='index', 
                                             columns=['val'])
feat_imp_rank = feat_imp_val.rank(method='max').rename(columns={"val": "rank"})
feat_imp = pd.concat([feat_imp_val, feat_imp_rank], axis=1, sort=False)

feat_imp
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 2))

for i in range(5):
    plt.subplot(1, 5, i+1)
    sns.scatterplot(x_train[:,i],y_train);
    plt.title('Feature ' + str(i))

plt.show()
plt.close()
````

### Gradient Boosting Trees
Is a chain of models that train on the residuals (errors) of the preceding model.

```
| Data |  y1_hat  |  m1_res  |  y2_hat  |  m2_res  |  y3_hat  |  m3_res  |
--------------------------------------------------------------------------
|   1  |    1.1   |    0.1   |   0.11   |   0.01   |  0.011   |   0.001  |
--------------------------------------------------------------------------
|   2  |    2.2   |    0.2   |   0.22   |   0.02   |  0.022   |   0.002  |
--------------------------------------------------------------------------
|   3  |    3.3   |    0.3   |   0.33   |   0.03   |  0.033   |   0.003  |
--------------------------------------------------------------------------
|   4  |    4.4   |    0.4   |   0.44   |   0.04   |  0.044   |   0.004  |
--------------------------------------------------------------------------```


````python
from sklearn.ensemble import GradientBoostingClassifier
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
gbrt = GradientBoostingClassifier(random_state=3, max_depth=2, verbose=1)
gbrt.fit(X_train, y_train)
print('Accuracy on training set: {:.3f}'.format(gbrt.score(X_train, y_train)))
print('Accuracy on test set:     {:.3f}'.format(gbrt.score(X_test, y_test)))
````

### Model Stacking
We add the predictions of several models (e.g. A, B, C) as features to the original data set, then run yet another model (e.g. D) on this 'enriched' data set.

### Interview Questions
_Can a single tree outperform a bagged model?_
No

_Can a single tree outperform a random forest?_
Unlikely, but can happen (e.g. bad luck when sampling or when we have few features

_Can a liner regression outperform a RF?_
Yes, if the problem is linear and the linear regression solves it perfectly.

_In RF, is sampling with replacement?_
With replacement

_What type of trees go into a RF?_
CART trees

_How does RF compare to DL?_
DL for images / text, RF for tabular data

_What is the bias / variance dilemma?_
It's typically a trade-off problem. More variance means overfitting (can be fixed by making the model less complex).

_What are disadvantages of RF?_
Does not work well on text. Large data sets create large trees?

_Does it matter if we have discrete or continuous variables for RF?_
No.
