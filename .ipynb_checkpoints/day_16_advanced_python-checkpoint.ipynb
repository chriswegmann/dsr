```python
# hwy = miles / gallon (highway)
# cty = miles / gallon (city)
```

```python
vehicles = pd.read_csv('http://www.gagolewski.com/resources/data/fueleconomy_vehicles.csv.gz',comment='#')
```

```python
# Given vehicles, convert the values in cty (city-) and hwy (highway-fuel economy -- given in mpg) to l/100 km.
# 1 mile = 1,60934 km
# 1 gallon = 3,78541l

vehicles["cty2"] = 235.215/vehicles["cty"]
vehicles["hwy2"] = 235.215/vehicles["hwy"]
```

```python
vehicles.sample(n=5, random_state=123)
```

```python
# Then add new columns z_cty and z_hwy, that give z-scores of cty and hwy (i.e., standardize these columns).
vehicles["z_cty"] = (vehicles.loc[:,"cty2"] - vehicles.loc[:,"cty2"].mean())/vehicles.loc[:,"cty2"].std()
vehicles["z_hwy"] = (vehicles.loc[:,"hwy2"] - vehicles.loc[:,"hwy2"].mean())/vehicles.loc[:,"hwy2"].std()
vehicles.sample(n=5, random_state=123)
```

```python
# Moreover, add two other columns z_cty_rel and z_hwy_rel, which denote the corresponding z-scores relative to 
# (grouped by) class.
```

```python
cty_std = vehicles.groupby("class").std()["cty2"].rename("cty_std")
cty_mean = vehicles.groupby("class").mean()["cty2"].rename("cty_mean")

hwy_std = vehicles.groupby("class").std()["hwy2"].rename("hwy_std")
hwy_mean = vehicles.groupby("class").mean()["hwy2"].rename("hwy_mean")

df = pd.concat([cty_std, cty_mean, hwy_std, hwy_mean], axis=1)
df.sample(n=5, random_state=123)
```

```python
vehicles = vehicles.join(df, on="class")
vehicles.sample(n=5, random_state=123)
```

```python
vehicles["z_cty_rel"] = (vehicles.loc[:,"cty2"] - vehicles.loc[:,"cty_mean"])/vehicles.loc[:,"cty_std"]
vehicles["z_hwy_rel"] = (vehicles.loc[:,"hwy2"] - vehicles.loc[:,"hwy_mean"])/vehicles.loc[:,"hwy_std"]
vehicles.sample(n=5, random_state=123)
```

```python
# elegant solution (above seems to be wrong, to be checked)
#vehicles["z_cty_rel"] = vehicles.groupby("class")[["cty2"]].transform(lambda x: (x-x.mean()/x.std()))

# when using transform, we should restrict ourselves to column(s)
```

```python
vehicles.sample(n=5, random_state=123)
```

```python
vehicles[["z_cty_rel", "z_hwy_rel"]] = vehicles.groupby("class")[["cty2", "hwy2"]].transform(lambda x: ((x-x.mean())/x.std()))
vehicles.sample(n=5, random_state=123)
```

```python
weather = pd.read_csv("http://www.gagolewski.com/resources/data/nycflights13_weather.csv.gz",comment="#")
```

```python
weather.head()
```

```python
weather["tempC"] = (weather["temp"]-32)*5/9
```

```python
weather_jfk = weather.query("origin=='JFK'").groupby(["year","month","day"])["tempC"].mean().reset_index()
```

```python
weather_jfk.head()
```

```python
weather_jfk = weather.query("origin=='JFK'").loc[:,["year","month","day","hour","tempC"]].reset_index(drop=True)
weather_jfk.head()
```

```python
what_we_have = pd.to_datetime(weather_jfk.loc[:,["year","month","day","hour"]])
all_hours    = pd.date_range("2013-01-01 00:00:00", periods=365*24, freq="H")
difference   = all_hours[~all_hours.isin(what_we_have)]
```

```python
len(difference)
```

```python
weather_jfk2 = pd.concat((weather_jfk, pd.DataFrame({
    "year":  difference.year,
    "month": difference.month,
    "day":   difference.day,
    "hour":  difference.hour,
    "tempC": np.nan
})), axis=0).sort_values(["year","month","day","hour"]).reset_index(drop=True)
weather_jfk2.head(7)
```

```python
weather_jfk2["tempC"] = weather_jfk2["tempC"].interpolate()
```

```python
weather_jfk = weather_jfk2.groupby(["year","month","day"])["tempC"].mean().reset_index()
```

```python
weather_jfk.head()
```

```python
weather_jfk.index = pd.to_datetime(weather_jfk.loc[:,["year","month","day"]])
weather_jfk["tempC"].plot();
```

```python
# Choose days with greater mean temperature than in the preceding day.
weather_jfk["tempC"].diff().head()
```

```python
weather_jfk["tempC"].head()
```

```python
weather_jfk.loc[weather_jfk["tempC"].diff()>0,:].head()
```

```python
weather_jfk.sort_values(["tempC"],ascending=False).head(5)
```

```python
weather_jfk.nlargest(5,["tempC"])
```
