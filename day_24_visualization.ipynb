```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import plotly as py
py.offline.init_notebook_mode(connected=True)
```

```python
table = pd.read_csv('https://raw.githubusercontent.com/chumo/Data2Serve/master/transition_clusters.csv')
```

```python
table.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(table.Xf, table.Yf, c=table.color)
ax.set_title('Random clusters')
```

```python
from plotly.tools import mpl_to_plotly
```

```python
plotly_dict = mpl_to_plotly(fig)
```

```python
py.offline.iplot(plotly_dict)
```

```python
trace = dict(
    type='scatter',
    x = table.Xf,
    y = table.Yf,
    marker = dict(
        color = table.color
    ),
    mode = 'markers',
)

my_data = [trace]

my_layout = dict(
    title = 'Random clusters',
    width = 500,
    height = 400,
)

fig = dict(
    data = my_data,
    layout = my_layout,
)
```

```python
py.offline.iplot(fig)
```

```python
py.offline.plot(fig, filename='day_24_pyplot_export.html')
```

```python
div_str = py.offline.plot(fig, output_type='div', include_plotlyjs=False)
div_str
# when including it on a web page: the d3 and plotly library need to be loaded
```

```python
trace = dict(
    type='histogram',
    x = table.Xf,
    nbinsx = 20,
)

my_data = [trace]

my_layout = dict(
    title = 'Random clusters',
    width = 500,
    height = 400,
    xaxis = dict(title = 'bins'),
    yaxis = dict(title = 'occurences'),
)

fig = dict(
    data = my_data,
    layout = my_layout,
)
py.offline.iplot(fig)
```

```python
trace1 = dict(
    type='histogram',
    name = 'Xf',
    x = table.Xf,
)

trace2 = dict(
    type='histogram',
    name = 'Xi',
    x = table.Xi,
)

my_data = [trace1, trace2]

my_layout = dict(
    title = 'Random clusters',
    width = 500,
    height = 400,
    xaxis = dict(title = 'bins'),
    yaxis = dict(title = 'occurences'),
)

fig = dict(
    data = my_data,
    layout = my_layout,
)
py.offline.iplot(fig)
```

```python
trace = dict(
    type= 'scatter',
    x= table.Xi,
    y= table.Yi,
    marker= dict(color=table.color),
    mode= 'markers',
)

data = [trace]

layout = dict(
    title= 'Random clusters',
    width= 1000,
    height= 1000,
    xaxis=dict(range=[0, 500]),
    yaxis=dict(range=[0, 500]),
    updatemenus= [dict(
        type= 'buttons',
        buttons= [{'label': 'Play', 'method': 'animate', 'args': [None]}],
    )]
)

frames = [
    dict(data=[{'x':table.Xi, 'y':table.Yi}]),
    dict(data=[{'x':table.Xf, 'y':table.Yf}]),
]

fig = dict(
    data=data,
    layout=layout,
    frames=frames,
)

py.offline.iplot(fig)
```
