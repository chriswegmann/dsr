```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/predictions.csv')
```

```python
model = 'predGbt'
```

```python
def calc_fpr_tpr(df, c):

    df_above_c = df.iloc[0:c]
    
    fpr = len(df_above_c[df_above_c['label']==0]) / len(df[df['label']==0])    
    tpr = len(df_above_c[df_above_c['label']==1]) / len(df[df['label']==1])
    
    return fpr, tpr
```

```python
df.sort_values(model, ascending=False, inplace=True)

fpr_all = []
tpr_all = []

for i in range(len(df)):
    fpr, tpr = calc_fpr_tpr(df, i+1)
    fpr_all.append(fpr)
    tpr_all.append(tpr)
```

```python
roc_fpr = pd.DataFrame(fpr_all).rename(index=str, columns={0: 'fpr'})
roc_tpr = pd.DataFrame(tpr_all).rename(index=str, columns={0: 'tpr'})
roc = pd.concat([roc_fpr, roc_tpr], axis=1)
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
ax = sns.lineplot(x="fpr", y="tpr", data=roc)
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, threshold = roc_curve(df['label'], df[model])
ax = sns.lineplot(x=fpr, y=tpr)
```
